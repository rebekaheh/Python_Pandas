```python
# Read with pandas--low_memory required to suppress errors about mixed data types
coders_df=pd.read_csv(r"C:\Users\RebekahDSK\Desktop\gwu-arl-data-pt-09-2020-u-c\02-Homework\04-Python-Pandas\03-Pandas-Project\Mini-Project-Part-2\Resources\2016-FCC-New-Coders-Survey-Data.csv")
coders_df
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
extracted_df= coders_df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
extracted_df.head(5)
```

```python
# Change reading 0 and 1 to No and Yes, respectively
#extracted_df = extracted_df.replace({0: "No", 1: "Yes"})
#extracted_df.head(5)
```

```python
# Change reading 0 and 1 to No and Yes, respectively, in the BootcampRecommend column
extracted_df["AttendedBootcamp"] = extracted_df["AttendedBootcamp"].replace({0: "No", 1 : "Yes"})
extracted_df["AttendedBootcamp"].value_counts()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = extracted_df.loc[extracted_df['AttendedBootcamp'] == "Yes"]
attended_bootcamp.head(5)
```

```python
# Determine which Bootcamps
boot_count = attended_bootcamp["BootcampName"].value_counts().count()
boot_signif_count = attended_bootcamp["BootcampName"].value_counts().loc[lambda x : x>25].count()

boot_signif_table = attended_bootcamp["BootcampName"].value_counts().loc[lambda x : x>25]
boot_signif_df = pd.DataFrame(boot_signif_table)
boot_signif_cvt_df = boot_signif_df.reset_index().rename(columns={"index": "Bootcamp","BootcampName" : "Students"})
boot_signif_cvt_df
```

```python
boot_signif_list = boot_signif_cvt_df["Bootcamp"].tolist()
print(boot_signif_list)
type(boot_signif_list)
```

```python
#NewBootcamp_df = Bootcampnames_df.count().sort_values("AttendedBootcamp", ascending=False).head()
#NewBootcamp_df = NewBootcamp_df[["AttendedBootcamp"]].head()
#NewBootcamp_df
```

```python
# Dataframe showing sig. # of attendees
something = attended_bootcamp[attended_bootcamp['BootcampName'].isin(boot_signif_list)]
something.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively, in the BootcampRecommend column
something["BootcampRecommend"] = something["BootcampRecommend"].replace({0: "No", 1 : "Yes"})
something["BootcampRecommend"].value_counts()
```

```python
# Create DataFrame of the attendees of Significant Bootcamps who recommended bootcamp
recommend_df = something[something["BootcampRecommend"] == "Yes"]
recommend_df.head()
```

```python
# Count how many attendees of each Significant Bootcamp would recommend the bootcamp
signif_boot_group = recommend_df.groupby(["BootcampName"]).count()
rec_signif_boot_group = signif_boot_group["BootcampRecommend"]
rec_signif_boot_group_df = pd.DataFrame(rec_signif_boot_group)

# rename columns
rec_signif_boot_group_df = rec_signif_boot_group_df.reset_index().rename(columns={"BootcampName": "Bootcamp","BootcampRecommend" : "Recommend"})

print(f"Number of People Who Attended Significant Bootcamps Who Also Recommend Those Bootcamps\n")
rec_signif_boot_group_df
```

```python
# Merge the two created data frames on the name of tbe bootcamp
# Merge two dataframes with inner join

rec_signif_boot_merge_df = pd.merge(boot_signif_cvt_df, rec_signif_boot_group_df, on="Bootcamp").reset_index()

# Add percentage column to merged dataframe
rec_signif_boot_merge_df["Percentage"] = (rec_signif_boot_merge_df["Recommend"] / rec_signif_boot_merge_df["Students"]) * 100

# Delete redundant "index" column
del rec_signif_boot_merge_df["index"]

# Improve formatting before outputting spreadsheet
rec_signif_boot_merge_df = rec_signif_boot_merge_df.style.format({"Percentage": "{:,.1f}%"})

rec_signif_boot_merge_df.hide_index()
```
