```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
coders_df=pd.read_csv(r"C:\Users\RebekahDSK\Desktop\gwu-arl-data-pt-09-2020-u-c\02-Homework\04-Python-Pandas\03-Pandas-Project\Mini-Project-Part-1\Resources\2016-FCC-New-Coders-Survey-Data.csv")
coders_df
```

```python
# Inspect all columns
coders_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
extracted_df= coders_df.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
extracted_df
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
extracted_df = extracted_df.replace({0: "No", 1: "Yes"})
extracted_df
```

```python
# Calculate total number of respondents in survey
total_respondents = len(extracted_df)
total_respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_bootcamp = extracted_df.loc[extracted_df['AttendedBootcamp'] == "Yes"]
attended_bootcamp.head()
```

```python
# Calculate average age of attendees
average_age = attended_bootcamp["Age"].mean()
print(average_age)
```

```python
# Calculate how many people attended a bootcamp
people_count = attended_bootcamp["AttendedBootcamp"].count()
print(people_count)
```

```python
# Calculate how many attendees hold degrees
school_degrees = (attended_bootcamp["SchoolDegree"].value_counts())
print(school_degrees)
```

```python
holds_degrees = attended_bootcamp["SchoolDegree"].loc[
    (attended_bootcamp["SchoolDegree"] != "some college credit, no degree") &
    (attended_bootcamp["SchoolDegree"] != "high school diploma or equivalent (GED)") &
    (attended_bootcamp["SchoolDegree"] != "trade, technical, or vocational training") &
    (attended_bootcamp["SchoolDegree"] != "some high school") &
    (attended_bootcamp["SchoolDegree"] != "no high school (secondary school)")
].count()

print(holds_degrees)
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender = attended_bootcamp["Gender"].count()
male = attended_bootcamp["Gender"].value_counts()["male"]
female = attended_bootcamp["Gender"].value_counts()["female"]
non_gender_specific = total_gender - male - female
print(f" Total: {total_gender}\n Male: {male}\n Female: {female}\n non_specific: {non_gender_specific}")
```

```python
total_attended = extracted_df['AttendedBootcamp'].count()
print(total_attended)
```

```python
# Calculate percentage of respondents who attended a bootcamp
attended_percentage = (people_count/total_respondents)*100
print(attended_percentage)
```

```python
# Calculate percentage of respondents belonging to each gender
male_percent = (male/total_gender)*100
female_percent = (female/total_gender)*100   
non_gender_specific = (non_gender_specific/total_gender)*100
print(f" The male % is {male_percent},\n the female % is {female_percent},\n and the non specific % is {non_gender_specific}")
```

```python
# Calculate percentage of attendees with a school degree
degree_percentage = (holds_degrees/people_count)*100
degree_percentage
```

```python
# Calculate average post-bootcamp salary
average_salary = (attended_bootcamp["BootcampPostSalary"].mean())
print(average_salary)
```

```python
# Create a new table consolodating above calculations
bootcamp_table = pd.DataFrame ({"Total Respondents": [total_respondents],
                                "Total Bootcamp Attendees": [people_count],
                                "Percent Attended Bootcamp": [attended_percentage],
                                "Average Age": [average_age],
                                "Percent Male": [male_percent],
                                "Percent Female": [female_percent],
                                "Percent Non Specific": [non_gender_specific],
                                "Percent with a Degree": [degree_percentage],
                                "Average Post Bootcamp Salary": [average_salary]})
bootcamp_table = bootcamp_table[["Total Respondents",
                                "Total Bootcamp Attendees",
                                "Percent Attended Bootcamp",
                                "Average Age",
                                "Percent Male",
                                "Percent Female",
                                "Percent Non Specific",
                                "Percent with a Degree",
                                "Average Post Bootcamp Salary"]]
bootcamp_table = bootcamp_table.round(2)

bootcamp_table
            
```

```python
# Improve formatting before outputting spreadsheet
bootcamp_table["Percent Male"] = bootcamp_table["Percent Male"].map(
    "{0:,.2f}%".format)
bootcamp_table["Percent Female"] = bootcamp_table["Percent Female"].map(
    "{0:,.2f}%".format)
bootcamp_table["Percent Non Specific"] = bootcamp_table["Percent Non Specific"].map(
    "{0:,.2f}%".format)
bootcamp_table["Percent with a Degree"] = bootcamp_table["Percent with a Degree"].map(
    "{0:,.2f}%".format)
bootcamp_table["Average Post Bootcamp Salary"] = bootcamp_table["Average Post Bootcamp Salary"].map(
    "${0:,.0f}".format)
bootcamp_table
```

```python
# Export to Excel
bootcamp_table.to_excel("BootcampOutputPart1.xlsx", index=False)
```
